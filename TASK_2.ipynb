```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


data=pd.read_csv("Unemployment in India.csv")
print(data.head())

data2=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
print(data2.head())

print(data2.isnull().sum())
data2.columns=['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employment','Estimated Labour Participation Rate','Region','longitude','latitude']
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
sns.heatmap(data2.corr())
plt.show()


data2.columns=["States","Date","Frequency","Estimated Unemployment Rate","Estimated Employment","Estimated Labour Participation Rate","Region","longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employment",hue="Region",data=data2)
plt.show()


unemployment=data2[["States","Region","Estimated Unemployment Rate"]]
figure=px.sunburst(unemployment, path=["Region","States"],
                   values="Estimated Unemployment Rate",
                   width=700,height=700,color_continuous_scale="RdY1Gn",
                   title="Unemployment Rate in India")
figure.show()
```
